```python
import random
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np
```

```python
N=1000 #number of data
m=5 #slope
n=10 #y intersection
Data=[]
while len(Data)<N:
    x = random.random()# random number btw 0 and 1
    y = m * x + n +3*(random.random()-0.5)# our target data
    Data.append((x,y))
```

```python
plt.scatter([d[0] for d in Data],[d[1] for d in Data])
```

```python
def J(theta1,theta0,Data):
    cost = 0
    for x,y in Data:
        cost += (y-theta1*x-theta0)**2
    return cost/len(Data)/2
```

```python
def gradJ(theta1,theta0,Data):
    grad0=0
    grad1=0
    for x,y in Data:
        grad1 += -1*(y-theta1*x-theta0)*(x)
        grad0 += -1*(y-theta1*x-theta0)
    return grad1/len(Data),grad0/len(Data)
```

```python
theta1=300
theta0=123
print(J(theta1,theta0,Data))
print(gradJ(theta1,theta0,Data))
```

```python
alpha=1# learning rate
```

```python
theta1
```

```python
theta0
```

```python
grad1,grad0 =gradJ(theta1,theta0,Data)
theta1 = theta1 - grad1 * alpha
theta0 = theta0 - grad0 * alpha
```

```python
theta1,theta0
```

```python
while True:#gradient descent
    grad1,grad0 =gradJ(theta1,theta0,Data)
    theta1 = theta1 - grad1 * alpha
    theta0 = theta0 - grad0 * alpha
    if grad1**2+grad0**2<0.00000001:
        break
```
